```python
def cap_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
    return df

df_cleaned = df.copy()

for col in numerical_features:
    df_cleaned = cap_outliers_iqr(df_cleaned, col)

print("Outliers capped for selected numerical features in 'df_cleaned'.")
```

```python
plt.figure(figsize=(20, 25))
for i, col in enumerate(numerical_features):
    plt.subplot(7, 3, i + 1)
    sns.boxplot(y=df_cleaned[col])
    plt.title(f'Box plot of {col} (Capped Outliers)')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
encoder = LabelEncoder()
scaler = StandardScaler()
```

```python
for col in categorical_features:
    df_preprocess[col] = encoder.fit_transform(df[col])

df_preprocess[categorical_features].head()
```

```python
df_processed = df_preprocess.copy()
X = df_processed.drop('Calories', axis=1)
y = df_processed['Calories']
```

```python
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
```

```python
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
print(f'df: {df_processed.shape}')
print(f'X_train: {X_train.shape}')
print(f'X_test: {X_test.shape}')
print(f'y_train: {y_train.shape}')
print(f'y_test: {y_test.shape}')
```

## Modeling

```python
lr = LinearRegression()
rf = RandomForestRegressor()
xgb = XGBRegressor()
lgbm = LGBMRegressor()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

lr = LinearRegression()
rf = RandomForestRegressor(random_state=42)
xgb = XGBRegressor(random_state=42)
lgbm = LGBMRegressor(random_state=42)

models = {
    'Linear Regression': lr,
    'Random Forest': rf,
    'XGBoost': xgb,
    'LightGBM': lgbm
}
n_features = 15
rfe_selectors = {}
results_rfe = {}
results_eval = {}

for name, model in models.items():
    rfe = RFE(model, n_features_to_select=n_features)
    rfe.fit(X_train, y_train)
    selected_features_mask = rfe.support_
    selected_features = X_train.columns[selected_features_mask]
    rfe_selectors[name] = rfe
    results_rfe[name] = selected_features

    X_train_selected = X_train[selected_features]
    X_test_selected = X_test[selected_features]


    model.fit(X_train_selected, y_train)


    y_pred = model.predict(X_test_selected)
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    results_eval[name] = {'R2 Score': r2, 'MAE': mae, 'RMSE': rmse}
```

```python
for name, features in results_rfe.items():
    print(f"\nModel: {name}")
    print("Fitur terpilih:", list(features))
```

```python
for name, metrics in results_eval.items():
    print(f"\nModel: {name}")
    print("R2 Score:", metrics['R2 Score'])
    print("MAE:", metrics['MAE'])
    print("RMSE:", metrics['RMSE'])

result_df = pd.DataFrame(results_eval).T
print("\nModel Evaluation Results:")
display(result_df)
```

```python
from sklearn.model_selection import GridSearchCV
import time

pram_grid_lr = {
    'fit_intercept': [True, False],
}

param_grid_rf = {
    'n_estimators': [50, 100],
    'max_depth': [10, 20],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt']
}

param_grid_xgb = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5],
    'learning_rate': [0.01, 0.1]
}

param_grid_lgbm = {
    'n_estimators': [100, 200],
    'num_leaves': [31, 63],
    'learning_rate': [0.01, 0.1]
}

grid_search_lr = GridSearchCV(lr, pram_grid_lr, cv=3, scoring='r2', n_jobs=-1)
grid_search_rf = GridSearchCV(rf, param_grid_rf, cv=3, scoring='r2', n_jobs=-1)
grid_search_xgb = GridSearchCV(xgb, param_grid_xgb, cv=3, scoring='r2', n_jobs=-1)
grid_search_lgbm = GridSearchCV(lgbm, param_grid_lgbm, cv=3, scoring='r2', n_jobs=-1)

tuned_models = {}
results_tuned_eval = {}

print("Tuning Linear Regression...")
start_time = time.time()
grid_search_lr.fit(X_train[results_rfe['Linear Regression']], y_train)
end_time = time.time()
training_time_lr = end_time - start_time
tuned_models['Linear Regression'] = grid_search_lr.best_estimator_
print(f"Best parameters for Linear Regression: {grid_search_lr.best_params_}")

y_pred_lr_tuned = tuned_models['Linear Regression'].predict(X_test[results_rfe['Linear Regression']])
r2_lr_tuned = r2_score(y_test, y_pred_lr_tuned)
mae_lr_tuned = mean_absolute_error(y_test, y_pred_lr_tuned)
rmse_lr_tuned = np.sqrt(mean_squared_error(y_test, y_pred_lr_tuned))
results_tuned_eval['Linear Regression'] = {'R2 Score': r2_lr_tuned, 'MAE': mae_lr_tuned, 'RMSE': rmse_lr_tuned, 'Training Time (s)': training_time_lr}

print("Tuning RandomForestRegressor...")
start_time = time.time()
grid_search_rf.fit(X_train[results_rfe['Random Forest']], y_train)
end_time = time.time()
training_time_rf = end_time - start_time
tuned_models['Random Forest'] = grid_search_rf.best_estimator_
print(f"Best parameters for Random Forest: {grid_search_rf.best_params_}")

y_pred_rf_tuned = tuned_models['Random Forest'].predict(X_test[results_rfe['Random Forest']])
r2_rf_tuned = r2_score(y_test, y_pred_rf_tuned)
mae_rf_tuned = mean_absolute_error(y_test, y_pred_rf_tuned)
rmse_rf_tuned = np.sqrt(mean_squared_error(y_test, y_pred_rf_tuned))
results_tuned_eval['Random Forest'] = {'R2 Score': r2_rf_tuned, 'MAE': mae_rf_tuned, 'RMSE': rmse_rf_tuned, 'Training Time (s)': training_time_rf}

print("Tuning XGBoost...")
start_time = time.time()
grid_search_xgb.fit(X_train[results_rfe['XGBoost']], y_train)
end_time = time.time()
training_time_xgb = end_time - start_time
tuned_models['XGBoost'] = grid_search_xgb.best_estimator_
print(f"Best parameters for XGBoost: {grid_search_xgb.best_params_}")

y_pred_xgb_tuned = tuned_models['XGBoost'].predict(X_test[results_rfe['XGBoost']])
r2_xgb_tuned = r2_score(y_test, y_pred_xgb_tuned)
mae_xgb_tuned = mean_absolute_error(y_test, y_pred_xgb_tuned)
rmse_xgb_tuned = np.sqrt(mean_squared_error(y_test, y_pred_xgb_tuned))
results_tuned_eval['XGBoost'] = {'R2 Score': r2_xgb_tuned, 'MAE': mae_xgb_tuned, 'RMSE': rmse_xgb_tuned, 'Training Time (s)': training_time_xgb}

print("Tuning LightGBM...")
start_time = time.time()
grid_search_lgbm.fit(X_train[results_rfe['LightGBM']], y_train)
end_time = time.time()
training_time_lgbm = end_time - start_time
tuned_models['LightGBM'] = grid_search_lgbm.best_estimator_
print(f"Best parameters for LightGBM: {grid_search_lgbm.best_params_}")

y_pred_lgbm_tuned = tuned_models['LightGBM'].predict(X_test[results_rfe['LightGBM']])
r2_lgbm_tuned = r2_score(y_test, y_pred_lgbm_tuned)
mae_lgbm_tuned = mean_absolute_error(y_test, y_pred_lgbm_tuned)
rmse_lgbm_tuned = np.sqrt(mean_squared_error(y_test, y_pred_lgbm_tuned))
results_tuned_eval['LightGBM'] = {'R2 Score': r2_lgbm_tuned, 'MAE': mae_lgbm_tuned, 'RMSE': rmse_lgbm_tuned, 'Training Time (s)': training_time_lgbm}


print("\nModel Evaluation Results After Tuning:")
best_results_df = pd.DataFrame(results_tuned_eval).T
display(best_results_df)
```
